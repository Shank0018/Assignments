```python
#Q.1:Check whether number is positive,negative or zero
n=int(input('Enter number :'))
if n>0:
    print('positive number ')
elif n==0:
    print('zero')
else:
    print('negative number')
```

```python
#Q.2 Check whether number is even or odd
n=int(input('enter any number '))
if n%2==0:
    print('even number')
else:
    print('odd number ')
    
```
